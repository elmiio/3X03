```python
import numpy as np
```

```python
# Constants
c = 3e5       #speed of light in km/s
H0 = 70       #Hubble constant in km/s/Mpc
D_H = c / H0  #Hubble distance in Mpc
 
#cosmological parameters (flat LCDM)
Om = 0.3
OL = 0.7
Ok = 0.0
```

```python
def E(z):
    return np.sqrt(Om * (1 + z)**3 + Ok * (1 + z)**2 + OL)
```

```python
def comoving_distance(z, N = 1000):
    if z == 0:
        return 0.0
    dz = z / N
    z_mids = (np.arange(N) + 0.5) * dz  #midpoints
    integral = np.sum(1.0 / E(z_mids)) * dz
    return D_H * integral
```

```python
DC_lens   = comoving_distance(0.5)
DC_source = comoving_distance(1.0)
 
print(f"D_C(z_L = 0.5) = {DC_lens:.1f} Mpc")
print(f"D_C(z_S = 1.0) = {DC_source:.1f} Mpc")
```
