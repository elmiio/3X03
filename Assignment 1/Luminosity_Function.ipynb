# import stuff

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define constants

```python
#absolute magnitude
M = 4.8

#solid angle
omega = 8.46e-6 #steradians

#stellar density
D = 1/25 #stars/pc^3
```

# data

```python
#magnitude bins
mags = np.arange(8, 20)

#number of stars per bin
n_stars = np.zeros_like(mags)
```

# calculate number of stars per bin

```python
for i, m in enumerate(mags):
    
    #distance for left edge of mag bin (m)
    d1 = 10**((m - M + 5)/5)
    
    #distance for right edge of mag bin (m + 1)
    d2 = 10**(((m + 1) - M + 5)/5)
    
    #volume
    V = (1/3) * omega * (d2**3 - d1**3)
    
    #number of stars
    N = V * D
    
    #store N in array
    n_stars[i] = N
```

# plot

```python
plt.figure(figsize = (10, 7))

plt.title('Luminosity Function Plot', fontsize = 15)

plt.plot(mags, n_stars, marker = 'o')

plt.xlabel('Apparent Magnitude [m]', fontsize = 12)
plt.ylabel('Number of Stars [N]', fontsize = 12)

plt.savefig('LumFun_Lin.png')
```

```python
plt.figure(figsize = (10, 7))

plt.title('Luminosity Function Plot', fontsize = 15)

plt.plot(mags, n_stars, marker = 'o')

plt.yscale('symlog')

plt.xlabel('Apparent Magnitude [m]', fontsize = 12)
plt.ylabel('Number of Stars [N]', fontsize = 12)

plt.savefig('LumFun_Lin.png')
```
